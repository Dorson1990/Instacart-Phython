```python
# Import libraries
import numpy as np
import pandas as pd
import os
import scipy
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Import the last updated merged data frame
pd.set_option('display.max_columns', None)
path = r'/Users/gomula/Documents/Instacart Basket Analysis 14.05.2023'
df_merged = pd.read_pickle(os.path.join(path, 'Data','Prepared Data', 'final_mergedv4.pkl'))
df_merged.head()
```

# Prepare visualization for final report

# Key question 1:

```python
# busiest days of the week
bar1 = df_merged['order_dow'].value_counts().sort_index().plot.bar(xlabel = 'Days of the week', ylabel = 'Order counts', title = 'Busiest days of the week')
```

```python
# save the chart
bar1.figure.savefig(os.path.join(path, 'Analysis', 'busydays.png'))
```

```python
#  busiest hours of the day
hist2 = df_merged['order_hour_of_day'].plot.hist(bins=24, ylabel = 'Order counts', title = 'Busiest hours of the day',  xlabel = 'Hours of the day' )
```

```python
# save the chart
hist2.figure.savefig(os.path.join(path, 'Analysis', 'busiesthours.png'))
```


# Key Question 2:

```python
#  price in different hours of the day 
line2 = sns.lineplot(data=df_merged, x='order_hour_of_day', y='prices')
```

```python
# save the chart
line2.figure.savefig(os.path.join(path, 'Analysis', 'pricehour.png'))
```

# Key Question 3:

```python
#  products price ranges
price_range = df_merged.groupby('product_id').agg({'price_range_loc' : 'max' })
price_range
```

```python
# drawing pie chart
explode = (0, 0.1, 0)
piechart1 = price_range.value_counts().plot.pie(autopct='%1.1f%%',
        shadow=True, startangle=90, explode = explode, title = None)
```

```python
# save the chart
piechart1.figure.savefig(os.path.join(path, 'Analysis', 'pricedist.png'))
```
